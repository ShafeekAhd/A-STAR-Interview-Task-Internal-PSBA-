```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import sklearn.externals as extjoblib
import joblib
```

```python
malware_data = pd.read_csv(r'C:\Users\Shafeek\OneDrive - Coventry University\Desktop\A-STAR\MalwareSamples10000.csv')
malware_data
```

```python
malware_data.values
```

```python
malware_data.dtypes
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data ['hasExe'] = malware_data['hasExe'].map(convert)
malware_data.head()
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data['hasZip'] = malware_data['hasZip'].map(convert)
malware_data.head()
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data['hasPDF'] = malware_data['hasPDF'].map(convert)
malware_data.head()
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data['hasDoc'] = malware_data['hasDoc'].map(convert)
malware_data.head()
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data['hasUnknown'] = malware_data['hasUnknown'].map(convert)
malware_data.head()
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data['hasURL'] = malware_data['hasURL'].map(convert)
malware_data.head()
```

```python
convert = {'Yes': 1, 'No': 0}
malware_data['isMalware'] = malware_data['isMalware'].map(convert)
malware_data.head()
```

```python
convert = {'com.au': 0, 'edu.au': 1, 'co.uk': 2, 'net.au': 3, 'com': 4, '.in': 5, 'net': 6, 'other':7}
malware_data['senderDomainSuffix'] = malware_data['senderDomainSuffix'].map(convert)
malware_data.head()
```

```python
malware_data
```

```python
X = malware_data.drop(columns = ['isMalware','specimenId'])
X
```

```python
y = malware_data['isMalware']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
model = DecisionTreeClassifier()
model.fit (X_train,y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
score = accuracy_score(y_test, predictions)
```

```python
score
```

```python
joblib.dump(model, 'Malware-pre-trained-prediction.joblib')
```
